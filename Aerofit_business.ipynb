```python
#Outliers detection using IQR method:
df['Income'].describe()
# To find outliers in any column we first need to boxplot but before that we need to find some points:
q1 = np.percentile(df['Income'],25)
#Lower quartile
q3 = np.percentile(df['Income'],75)
#Upper quartile
print('q1=',q1)
print('q3=',q3)
```

```python
#To find upper and lower boundary we need to calculate IQR(Inter Quartile Range)
IQR = q3 - q1
IQR
```

```python
upper_bound = q3+1.5*IQR
lower_bound = q1-1.5*IQR
print("upper_bound=",upper_bound)
print("lower_bound=",lower_bound)
print('Median=',df['Income'].median())
```

```python
sns.boxplot(data = df,x='Income')
plt.show()
```

```python
def handle_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    IQR = q3-q1
    lower_bound = q1 - 1.5 *IQR
    upper_bound = q3 + 1.5 *IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
```

```python
columns_with_outliers = ['Income','Miles','Age','Usage']
df_cleaned = df.copy()
for col in columns_with_outliers:
    df_cleaned = handle_outliers(df_cleaned, col)
original_shape = df.shape
cleaned_shape = df_cleaned.shape
```

```python
#Clipping the data between 5th and 95th percentile
```

```python
def clip_data(data):
    lower_bound = np.percentile(data,5)
    upper_bound = np.percentile(data,95)
    clipped_data = np.clip(data,lower_bound,upper_bound)
    return clipped_data
    
```

```python
Continuous_variable_columns =['Age','Miles','Usage','Income']
for var in Continuous_variable_columns:
    clipped_data = clip_data(df[var])
    print(f'original{var}:\n',df[var])
    print(f'\nclipped{var}:\n',clipped_data)
    df[var+"_clipped"]= clipped_data
print(df)
```
